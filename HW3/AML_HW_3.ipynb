# Question 2 : t-SNE (7 pts)

Read this [article](https://erdem.pl/2020/04/t-sne-clearly-explained) on t-SNE, a dimensionality reduction technique for visualization and explain it in your own words in one or two paragraphs.

#### Answer:

t-distributed Stochastic Neighbor Embedding (t-SNE) is a dimensionality reduction technique similar to Principal Component Analysis (PCA). However, unlike PCA, t-SNE uses nonlinear dimensionality reduction to separate data that cannot be separated by a straight (linear) line. The algorithm works by first creating a probability distribution using a distance measurement. We can think of these similarity measurements as the conditional probability that a point in one cluster would belong to the same cluster as another data point. Given that clusters often have different variances, the algorithm uses a measurement known as perplexity, which gives a 'target' amount of neighbors for the data point being examined.

Next, a low-dimensional space is created with the same number of data points as in the original space. To account for the crowding problem of Gaussian distributions, we use a Student t-distribution for its 'fatter' tails. This makes it easier to distiguish the y-value for each data point being represented in the distribution. Finally, Kullback-Leibler divergence is used between the conditional probabilities to optimize the t-SNE distribution. A gradient is calculated for each point and describes the strength of the 'pull' towards the centroid of each cluster. 

# Question 3: PCA (8 pts)

Explain the principle of Principal Component Analysis algorithm and in what sense is the selection of the eigenvectors to represent the data is an optimal choice.  How do you reconstruct (a noisy version of) the original data from the eigenvectors and the loadings?



# Answer:

Principal Component Analysis 'PCA' is a method that is used to reduce the dimensions of a large set of data through transforming a set of variables in to a smaller set while also trying to keep as much information as possible from the original larger set. In other words, PCA attempts to retain as much information as possible with the fewest number of variables / dimensions. PCA is performed through obtaining / realizing the 'principle components' in the initial large data set and getting rid of all the other variables in which do not provide much relevance when conveying / obtaining as much information as possible.

Eigenvectors in PCA can be thought of as the 'columns'/components in the data and each eigenvector provides and conveys information about the initial large data set. In order to optimally select the eigenvectors, we need to decide the fewest eigenvectors that portray and maintain the most data pertaining to the original set.

In order to reconstruct a noisy version of the original data, we can multiply each principal component score by the transpose of the eigenvector matrix.

e) i)**(2pts)** Calculate 95% confidence intervals for SalePrice with a sample size of 100. 

ii)**(3pts)** Calculate 95 % confidence intervals for 100 different trials with a sample size of 500. Plot the confidence intervals and interpret how it captures the population mean.

````python
np.random.seed(10)
sample_size = 100
sample = np.random.choice(a= data_dm['SalePrice'],
                          size = sample_size)
sample_mean = sample.mean()
pop_mean = data_dm.SalePrice.mean()

#Get the critical Z value, 1.96 corresponds to 95% CI
z_critical = 1.96

#Get population standard deviation
pop_stdev = data_dm.SalePrice.std()

#margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

#confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  
                       
#Print confidence interval and true mean value
print('95% Confidence Interval: ', confidence_interval)
print('True Mean: ', pop_mean)
````

````python
np.random.seed(12)

#define the sample size
sample_size = 500

intervals = []
sample_means = []
pop_mean = data_dm.SalePrice.mean()

for sample in range(100):
    sample = np.random.choice(a= data_dm['SalePrice'], size = sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

    #compute z critical value
    z_critical = 1.96
    
    #compute population std dev        
    pop_stdev = data_dm.SalePrice.std()
    
    margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
    
    confidence_interval = (sample_mean - margin_of_error,
                           sample_mean + margin_of_error) 
    
    intervals.append(confidence_interval)


plt.figure(figsize=(13, 9))
plt.errorbar(x=np.arange(0.1, 100, 1),
             y=sample_means, 
             yerr=[(top-bot)/2 for bot,top in intervals],
             fmt='o')

plt.hlines(xmin=0, xmax=100,
           y=data['SalePrice'].mean(), 
           linewidth=2.0,
           color="red")
plt.title('Confidence Intervals for 100 Trials', fontsize = 20)
plt.show()
````

### Solution:

Each point on the graph represents the sample mean of a single trial out of one hundred. The margin of error for each sample was calculated by multiplying the standard error by the z-score value. Adding and subtracting this value from each sample mean gives us our 95% confidence interval. We can see in the graph that nearly every sample's confidence interval captures the true mean of the population (represented by the horizonal red line), except for 3 samples. This is not surprising given that calculating the 95% confidence interval means we are 95% confident that the true population mean lies within the interval. Although variation among each sample mean is relatively large, the confidence intervals gives us an idea of where the true mean is within the population.
